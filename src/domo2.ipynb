```python
#4 Create a mock agent for demonstration with all imports in same cell
import sys
import os
sys.path.append('src')

from tools.destination_research import research_destination
from tools.itinerary_builder import build_daily_itinerary
from tools.budget_calculator import calculate_budget_breakdown, generate_budget_chart
from tools.weather_checker import get_seasonal_weather
from tools.map_visualizer import generate_itinerary_map

class MockAgent:
    def __init__(self):
        self.tools = [
            research_destination,
            build_daily_itinerary,
            calculate_budget_breakdown,
            get_seasonal_weather,
            generate_itinerary_map,
            generate_budget_chart
        ]
    
    def plan_trip(self, destination, duration, budget, style="mixed", travelers=1, interests=None, month="May"):
        print(f"ğŸš€ Planning {duration}-day {style} trip to {destination}...")
        
        # Simulate LangGraph workflow steps
        steps = [
            "ğŸ” Researching destination...",
            "ğŸ“… Building itinerary...", 
            "ğŸ’° Calculating budget...",
            "ğŸŒ¤ï¸ Checking weather...",
            "ğŸ—ºï¸ Generating visualizations...",
            "ğŸ“ Creating final report..."
        ]
        
        for step in steps:
            print(step)
        
        # Execute tools sequentially (simulating LangGraph)
        dest_info = research_destination(destination, interests or ["sightseeing"])
        itinerary = build_daily_itinerary(destination, duration, style, dest_info['attractions'][:4])
        budget_data = calculate_budget_breakdown(destination, duration, budget, style, travelers)
        weather_info = get_seasonal_weather(destination, month)
        itinerary_map = generate_itinerary_map(destination, itinerary, dest_info['attractions'][:3])
        budget_viz = generate_budget_chart(budget_data['budget_breakdown'])
        
        return {
            'destination': destination,
            'itinerary': itinerary,
            'budget': budget_data,
            'weather': weather_info,
            'map': itinerary_map,
            'budget_chart': budget_viz,
            'attractions': dest_info['attractions'][:5]
        }

print("âœ… Mock Agent created for demonstration!")
mock_agent = MockAgent()
```

```python
#cell 5 
print("ğŸ‡«ğŸ‡· DEMO 1: Paris Cultural Trip")
print("=" * 40)

paris_result = mock_agent.plan_trip(
    destination="Paris",
    duration=4,
    budget=2000,
    style="cultural",
    travelers=2,
    interests=["museums", "history", "food"],
    month="May"
)

print(f"\nâœ… Paris Trip Planned!")
print(f"ğŸ“ {len(paris_result['attractions'])} attractions researched")
print(f"ğŸ“… {len(paris_result['itinerary'])} day itinerary built")
print(f"ğŸ’° ${paris_result['budget']['total_budget']} budget allocated")
print(f"ğŸŒ¤ï¸ {paris_result['weather']['temperature']} weather expected")
```
